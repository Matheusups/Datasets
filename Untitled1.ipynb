```
import numpy as np

import pandas as pd
df = pd.read_csv("/prouni2018.csv", index_col=0)
df.query('turno == "Curso a Distância"').head(5000)
#Query me indica os alunos correspondentes aos cursos EAD
```

```
df.query('mensalidade <= 300').head(5000)
#Alunos que pagam a mensalidade inferior ou igual a 300 reais
```

```
df.groupby(by='turno').size()
#Analise quantitativa em relação aos turnos 
```

```
df.groupby(['universidade_nome','nome','turno']).agg({'mensalidade': np.mean, 'cidade_busca':np.size}).head(5000)
#Analise da média do valor da mensalidade paga por Turno, Curso e Universidade
```
